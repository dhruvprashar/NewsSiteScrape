```python
import urllib.request,sys,time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait       
from selenium.webdriver.common.by import By       
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome(r"C:\Users\prash\Downloads\chromedriver_w\chromedriver")
driver
```

```python
driver.get("https://on-demand.eastview.com/browse/issues/515/2008/udb/1")
driver.maximize_window()
```

```python
date = driver.find_element("link text","No. 8-5")
date
```

```python
# date = driver.find_element("link text","No. 8-6")
# date.click()
frame=[]
for d in range(5,15):
    da="No. 8-" + str(d)
#     print(da)
    try:
        date = driver.find_element("link text",da)
        date.click()
        while True:
            try:
                data = BeautifulSoup(driver.page_source,'html.parser')
        #         Hls = data.find_elements("")
        #         print(data) 
                table= data.find("tbody",attrs= {'id':'sourcesArticles-tableBody'})
                hdl= table.find_all("a",attrs={'class':'Link'})
                ddate=table.find_all("span",attrs={'class':'Text-noWrap'})
                articles = table.find_all("td", attrs={'class':'Sidebar-excerpts'})
        #         print(ddate)
        #         print(hdl)
        #         print(articles)
                for i in range(len(hdl)):
                    headline=hdl[i].text.strip()
                    article=articles[i].find("p").text.strip()
                    dat=ddate[i].text.strip()
                    print(headline)
                    print(article)
                    print(dat)

                    frame.append((headline,article,dat))
        #         now we find the next button and click it to change page
                nextButton = driver.find_element("class name","Pagination-arrow--right")
                nextButton.click()
            except NoSuchElementException:
        #         if the nextButton doesnot exist we revert back to 2008 page and then go for next date that we wish to scrape
                driver.get("https://on-demand.eastview.com/browse/issues/515/2008/udb/1")
                break

        
    except NoSuchElementException:
        print("Date Not found")
        
```

```python

allnews=pd.DataFrame(frame,columns=['Headline','Article','Date'])
allnews.head()
```

```python
allnews.to_csv("Tass2008.csv", index= False, encoding= 'utf-8')
df=pd.read_csv("Tass2008.csv", encoding= 'utf-8')
df.head()
```
